<font color="A91D3A" size="5px" face="Inter"> <strong>5. Manipulation des données : </strong> </font> <br>
<br>
<font color="black" size="4px" face="Inter">
 <strong>  -Top 5 des meilleurs categories:<br></strong>
 <strong>  -Nous considérons que la meilleure catégorie est celle qui est répétée le plus grand nombre de fois.<br></strong>
    Nous avons remarqué que les cellules de la colonne "listed_in" contiennent plus d'une catégorie séparée par ','.
Ce code divise chaque cellule par virgule en utilisant la méthode str.split(',') pour créer une liste de catégories individuelles pour chaque ligne. Ensuite, il applique la fonction pd.Series() pour transformer chaque liste de catégories en une ligne séparée, ce qui donne un nouveau df avec plusieurs lignes pour chaque ligne d'origine.<br>
      Ensuite, la méthode stack() est utilisée pour faire pivoter les colonnes du nouveau df en une seule colonne, créant ainsi une série de toutes les catégories.<br>

 La méthode str.strip() pour supprimer tous les espaces.<br>
a méthode value_counts() pour compter chaque catégorie unique dans la série,elle donne le nombre de fois de chaque catégorie  dans la colonne "listed_in".
  
    
    
</font><br>

```python
category= df['listed_in'].str.split(',').apply(pd.Series).stack().str.strip().value_counts()
```

```python
category
```

```python
category.head(5)
```

<hr>

<font color="black" size="4px" face="Inter">
 <strong>  -Top 5 des meilleurs Realisateur:<br></strong>
 <strong>  -Nous considérons que la meilleure Realisateur est celle qui est répétée le plusieurs fois.<br></strong>
    Ce code va compter le nombre d'occurrences de chaque réalisateur dans la colonne "director" en utilisant la méthode          value_counts(), puis il retournera les 5 meilleurs réalisateurs avec les nombres les plus élevés en utilisant la méthode head().
</font>

```python
directors = df['director'].value_counts().head(5)
print(directors)
```

<hr>
<font color="black" size="4px" face="Inter">
Nous avons remarqué que les cellules NaN de la colonne 'director' que nous avons remplacées par 'Unknown' ont le plus grand nombre d'occurrences, nous devons donc filtrer et ne laisser que les cellules qui ne contiennent pas 'Unknown'.
</font>

```python
# les lignes avec 'Unknown' comme réalisateur
director = df[df['director'] != 'Unknown']

# nombre d'occurrences de chaque réalisateur
nb_director = director['director'].value_counts()
directors = nb_director.head(5)
print(directors)
```

-<font color="EEEEEE">['cast'].str.split(',')</font>:  "split" pour chaque valeur de la colonne "cast" et divise chaque chaîne de caractères en une liste d'acteurs séparés par ",".<br>

<font color="black" size="4px" face="Inter">
pour utiliser le modèle pré-entraîné pour effectuer des tâches de traitement de langage naturel telles que l'analyse syntaxique, dans la variable"nlp"</font><br> 

```python
nlp=spacy.load('en_core_web_lg')
```

<font color="black" size="4px" face="Inter">
Pour tester : <br>
On cree l'objet "d1" qui contient le résultat de l'analyse de la chaîne de caractères "horror" par le modèle de langage "nlp". <br>
"d2" qui contient le résultat de l'analyse de la phrase "The movie was so terrifying that it made me cover my eyes and scream in fear." par le modèle de langage "nlp". <br>
la méthode "similarity()" calcule la similarité sémantique entre ces deux d1 et d2.(retourn une valeur entre 0 et 1)</font><br> 

```python
d1=nlp("horror")
d2=nlp("The movie was so terrifying that it made me cover my eyes and scream in fear.")
```

```python
d1.similarity(d2)
```

<font color="black" size="4px" face="Inter">
Ce code calcule la similarité sémantique entre d1= "horror" et chaque élément de la colonne 'description' de "df".
Pour chaque élément de la colonne 'description',"nlp(s)" est utilisée pour créer un objet qui contient les informations d'analyse de la phrase "s" (la phrase "s" chaque phrase de la colonne 'description').la méthode "similarity()" est utilisée pour calculer la similarité sémantique entre "d1" et la phrase "x". <br>
<font color="green">nv_df = df.copy()</font>: crée une copie de "df" et la stocke dans une nouvelle variable "nv_df".<br>
<font color="green">nv_df['similarity_score'] = similarity_score</font> :pour ajoute une nouvelle colonne 'similarity_score' au la copy de df "nv_df".<br>
<font color="green">display(nv_df)</font> : pour afficher la nouvelle df de la bibliothèque Pandas de manière formatée. display() est une fonction spécifique à jupyter notebook.
 </font><br> 

```python
similarity_score = [d1.similarity(nlp(s)) for s in df['description']]

nv_df = df.copy()
nv_df['similarity_score'] = similarity_score

display(nv_df)
```

```python
#la valeur maximale de la colonne 'similarity_score' de "nv_df"
max_similarity = nv_df['similarity_score'].max()
print(max_similarity)
```

```python
#sélectionne les lignes de "nv_df" où la valeur de la colonne 'similarity_score' est égale à la valeur maximale 'max_similarity'
most_similar = nv_df[nv_df['similarity_score'] == max_similarity]
display(most_similar)
```

<font color="A91D3A" size="5px" face="Inter"> <strong>7. Netflix se concentre-il davantage sur les séries télévisées que sur les films ces derniéres années?  </strong> </font> <br>

<font color="black" size="4px" face="Inter">- On considère que "les dernières années" correspondent aux "5 dernières années".</font> <br>

```python
#la plus grande année de colonne 'release_year'
max_R_year=df['release_year'].max()
print(max_R_year)
```

```python
#les 5 dernières années par rapport la plus grande année
five_y=max_R_year-5
print(five_y)
```

```python
#le nombre de films et de séries TV dans "df"
types = df['type'].value_counts()
print(types)
```

```python
#Selectionner les films dont la date de réalisation est comprise entre l'année maximale et l'année maximale diminuée de 5 ans
selected_movies = df[(df['type'] == 'Movie') & (df['release_year'] >= five_y) & (df['release_year'] <= max_R_year)]
#Selectionner les séries TV dont la date de réalisation est comprise entre l'année maximale et l'année maximale diminuée de 5ans
selected_tvshow = df[(df['type'] == 'TV Show') & (df['release_year'] >= five_y) & (df['release_year'] <= max_R_year)]
if len(selected_movies) > len(selected_tvshow):
    print("Netflix se concentre sur les films des 5 dernières années")
else:
    print("Netflix se concentre sur les TV Shows des 5 dernières années")
```

<font color="A91D3A" size="5px" face="Inter"> <strong>8. Diagramme Circulaire :  </strong> <br></font><br>
<font color="black" size="4px">Les graphes sont très utiles en analyse de données , ils permettent de visualiser les données et de les présenter d'une manière plus facile à comprendre.</font>

```python
rat = df['rating'].value_counts()
print(rat)
```

 <hr>
 <font color="black" size="4px" face="Inter">
    <strong>plt</strong> : est un alias pour le module pyplot de la bibliothèque matplotlib en Python. <br>
 <strong>pie()</strong> : est la fonction de la bibliothèque matplotlib pour créer le graphique.<br>
 <strong>rat.values</strong> : contient les valeurs (comptages) pour chaque catégorie de classement (df['rating']).<br>
 <strong>rat.index</strong> : contient les catégories de df['rating'].<br>
 <strong>labels</strong> :est Le paramètre utilisé pour étiqueter chaque tranche du graphique avec les catégories de classement (df['rating']). <br><strong>autopct</strong>: est le paramètre formate les valeurs en pourcentage affichées sur chaque tranche.<br>
  <strong>plt.title()</strong> : pour définir le titre du graphique.<br>
  <strong>plt.show()</strong> : pour afficher le graphique.
    </font> <br>
 

```python
plt.pie(rat.values, labels=rat.index, autopct='%1.1f%%')
plt.title("Répartition du contenu par classement")
plt.show()
```
